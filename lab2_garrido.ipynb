```python
# importar librerías
import pandas as pd
import requests
from io import StringIO

'''
Para no descargar el csv y cargarlo en colba decidí utilizar requests
y cargarlo directamente desde la URL'''
url = 'https://datos.cdmx.gob.mx/dataset/c9e96ab4-c127-4ee1-a222-bfa2b5d759de/resource/b1926485-e512-4c4f-a03a-5d489bce8740/download/atenciones_abogadas_de_las_mujeres.csv'
response = requests.get(url)

data = StringIO(response.text)
df = pd.read_csv(data)
df
```

# 1. Verificar lectura de archivos CSV

```python
# Obtener el número de registros
registros  = df.shape[0]
print(registros)
```

# 4. Obtener el rango de fechas

```python
df["fecha"] = pd.to_datetime(df["fecha"], format="%Y-%m-%d")

fecha_inicial = df["fecha"].min()
fecha_final = df["fecha"].max()

print(f"El rango de fecha es del {fecha_inicial} al {fecha_final}")
```

# 5. Obtener la media aritmética

```python
media = df["usuaria_edad"].mean()
print(media)
```

# 6. Obtener la moda

```python
moda = df["usuaria_edad"].mode()
print(moda)
```
